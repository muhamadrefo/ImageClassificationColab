```
#Importing Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from datetime import datetime
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, precision_recall_fscore_support
from sklearn.metrics import confusion_matrix as cm
from sklearn import model_selection
from sklearn.model_selection import train_test_split as tt, GridSearchCV
from sklearn.metrics import classification_report
%matplotlib inline

%cd /content/drive/My Drive/Batik/SVM/
%ls
```

```
dTrain = pd.read_csv('newDataTraining.csv')
dValid = pd.read_csv('newDataTraining.csv')
x = dTrain.iloc[:, 1:6].values
y = dTrain.iloc[:, 7].values
valDat = dValid.iloc[:, 1:6].values
valTarget = dValid.iloc[:, 7].values
Rans = 100
```

```
def linear():
  startTime = datetime.now()
  model = SVC(kernel='linear')
  with open('RsLinear.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['RandomState', 'Accuracy Training', 'Accuracy Validation'])
    for rs in range(int(Rans)):
      x_train, x_test, y_train, y_test = tt(x,y, test_size=0.2, random_state=rs+1)
      model.fit(x_train, y_train)
      y_pred = model.predict(x_test)
      valPred = model.predict(valDat)
      accT = (np.sum(y_pred == y_test)/ y_test.size)*100
      accV = (np.sum(valPred == valTarget)/ valTarget.size)*100
      writer.writerow([str(rs+1),str(accT),str(accV)])
  print('Time Processing  : ', (datetime.now()-startTime))

linear()
showLin = pd.read_csv('RsLinear.csv')
showLin
```

```
def pandas_classification_report(y_true, y_pred):
        metrics_summary = precision_recall_fscore_support(
                y_true=y_true, 
                y_pred=y_pred)

        avg = list(precision_recall_fscore_support(
                y_true=y_true, 
                y_pred=y_pred,
                average='weighted'))

        metrics_sum_index = ['precision', 'recall', 'f1-score', 'support']

        class_report_df = pd.DataFrame(list(metrics_summary),index=metrics_sum_index)
        support = class_report_df.loc['support']
        total = support.sum() 
        avg[-1] = total
        
        # class_report_df = pd.DataFrame(index=metrics_sum_index)
        class_report_df['avg / total'] = avg

        return class_report_df.T

def linear2():
  startTime = datetime.now()
  model = SVC(kernel='linear')
  with open('RsLinear2.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['RandomState', 'Accuracy Training', 'Accuracy Validation', 'Clas Report'])
    for rs in range(int(Rans)):
      x_train, x_test, y_train, y_test = tt(x,y, test_size=0.2, random_state=rs+1)
      model.fit(x_train, y_train)
      y_pred = model.predict(x_test)
      valPred = model.predict(valDat)
      accT = (np.sum(y_pred == y_test)/ y_test.size)*100
      accV = (np.sum(valPred == valTarget)/ valTarget.size)*100
      cls_rpt = classification_report(y_test, y_pred)
      class_report = pandas_classification_report(valTarget, valPred)
      writer.writerow([str(rs+1),str(accT),str(accV), str(cls_rpt)])

  class_report.to_csv('RsTryLinear2.csv', header=True)
  print('Time Processing  : ', (datetime.now()-startTime))

linear2()
showLin2 = pd.read_csv('RsLinear2.csv')
showLin2
```
